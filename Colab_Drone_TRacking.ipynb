```python
import cv2
import os

# Input video file path
video_path = '/content/drive/MyDrive/Drone/Drone_test.mp4'  # Replace with the path to your video file

# Output directory where images will be saved
output_dir = '/content/drive/MyDrive/Drone/runs/Images_dir'  # Replace with the desired output directory

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Open the video file
cap = cv2.VideoCapture(video_path)

# Loop through the video frames and save them as images
frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Save the frame as an image
    image_name = f"frame_{frame_count:04d}.jpg"  # You can customize the image naming scheme
    image_path = os.path.join(output_dir, image_name)
    cv2.imwrite(image_path, frame)

    frame_count += 1

# Release the video capture object
cap.release()

print(f"Extracted {frame_count} frames and saved them in {output_dir}")
```
